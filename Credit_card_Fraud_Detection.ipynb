# Importing the Dependencies

```python
import numpy as np #for creating arrays
import pandas as pd # to read csv file
from sklearn.model_selection import train_test_split # for splitting data for training and testing
from sklearn.linear_model import LogisticRegression # Using Linear Regression Model
from sklearn.metrics import accuracy_score # for testing the accuracy of our model
from micromlgen import port
from glob import glob
from os.path import basename
from sklearn.ensemble import RandomForestClassifier
```

# Now importing the credit card dataset

```python
credit_card_data = pd.read_csv('/content/drive/MyDrive/fraudTrain.csv')
credit_card_data.head()
```

# Now we are going to check the info

```python
credit_card_data.info()
```

# Checking out normal and fraud transactions

```python
credit_card_data['is_fraud'].value_counts()
```

```python
# balancing the dataset
legit = credit_card_data[credit_card_data.is_fraud == 0]
fraud = credit_card_data[credit_card_data.is_fraud == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

# Now checking the statistical measurement of our data

```python
legit.amt.describe()
```

```python
fraud.amt.describe()
```

# Now we need to equalize our dataset of legit and fraud

```python
# Compare the values for both transactions
credit_card_data.groupby('is_fraud').mean()
```

# Downsampling for equalization

```python
legit_new = legit.sample(n = 7506)
```

# Now concatenating two dataframes to which we are concerned

```python
new_dataset = pd.concat([legit_new,fraud],axis = 0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
# Now checking for new dataset
new_dataset.groupby('is_fraud').mean()
```

# Now we will separate features and set target variable

```python
X = new_dataset.drop(columns='is_fraud',axis = 0)
Y = new_dataset['is_fraud']
```

```python
print(X)
```

```python
print(Y)
```

# Now comes the training and testing




```python
# Setting data for testing and traiuning
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Training our model using Logistic Regression

```python
from sklearn.preprocessing import OneHotEncoder

# Assuming X_train contains categorical features
encoder = OneHotEncoder()
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.fit_transform(X_test)

model = LogisticRegression()
model.fit(X_train_encoded, Y_train)
```

# Model Evaluation

Accuracy

```python
model_prediction = model.predict(X_train_encoded)
model_accuracy = accuracy_score(model_prediction,Y_train)

print("Training data accuracy:", model_accuracy)
```
